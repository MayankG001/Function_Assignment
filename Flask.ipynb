**Q3. What is App routing in Flask? Why do we use app routes?**
***

**In Flask**, app routing is the process of mapping a URL to a specific function that should be executed when a user visits that URL.
Flask app routes are defined using the @app.route decorator.
### For example:



In this example, the `/` URL is mapped to the index() function, which returns the string 'Hello, World!' when a user visits that URL.

App routes are useful for building web applications because they allow you to define the behavior of your application in response to user requests. You can use app routes to define the URLs that your application will respond to, and the functions that should be executed when those URLs are visited.

**Overall, app routing is a fundamental aspect of building web applications in Flask, and app routes allow you to define the behavior of your application in a clear and concise way.**

**Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.**
***

In Flask, the url_for() function is used for URL building. This function is particularly useful because it allows you to dynamically build URLs for your application based on the name of the view function and any arguments to be passed to it. This ensures that you avoid hard-coding URLs and makes your application more maintainable.The `url_for()` function is used in Flask for URL building. It takes the name of a view function as its first argument, and any number of keyword arguments that correspond to the variable parts of the URL. The function returns a URL that can be used to access the specified view function.


```python
from flask import Flask, url_for, redirect

app = Flask(__name__)

@app.route('/')
def index():
    return "Welcome to the Home Page!"

@app.route('/about')
def about():
    return "Welcome to the About Page!"

@app.route('/user/<username>')
def profile(username):
    return f"User: {username}"

@app.route('/redirect_to_about')
def redirect_to_about():
    # Using url_for to build the URL for the 'about' endpoint
    return redirect(url_for('about'))

@app.route('/redirect_to_profile')
def redirect_to_profile():
    # Using url_for to build the URL for the 'profile' endpoint with an argument
    return redirect(url_for('profile', username='JohnDoe'))

if __name__ == '__main__':
    app.run(debug=True)
```

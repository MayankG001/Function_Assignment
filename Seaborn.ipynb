Q1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.


1. Line Plot: Used for visualizing data points over a continuous interval or time span. It is useful for trend analysis.
2. Bar Plot: Used to show the relationship between a categorical variable and a continuous variable. It is useful for comparing different categories.
3. Box Plot: Used for visualizing the distribution of data based on a five-number summary. It is useful for identifying outliers and understanding data spread.
4. Histogram: Used for visualizing the frequency distribution of a continuous variable. It is useful for understanding the distribution of data.
5. Heatmap: Used for visualizing data in a matrix form where individual values are represented as colors. It is useful for identifying patterns and correlations in data.

Q2: Load the "fmri" dataset and plot a line plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Line plot using timepoint and signal for different events and regions
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
plt.title('FMRI Dataset Line Plot')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

Note: pclass, age, and fare are columns in the titanic dataset.

```python
# Load the titanic dataset
titanic = sns.load_dataset("titanic")

# Box plot using pclass and age
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(x="pclass", y="age", data=titanic)
plt.title('Box Plot of Age by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Age')

# Box plot using pclass and fare
plt.subplot(1, 2, 2)
sns.boxplot(x="pclass", y="fare", data=titanic)
plt.title('Box Plot of Fare by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Fare')

plt.tight_layout()
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Histogram for the price column with hue parameter for the cut column
sns.histplot(data=diamonds, x="price", hue="cut", multiple="stack")
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
# Load the iris dataset
iris = sns.load_dataset("iris")

# Pair plot with hue parameter for the species column
sns.pairplot(iris, hue="species")
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
# Load the flights dataset
flights = sns.load_dataset("flights")

# Pivot the dataset to create a matrix suitable for a heatmap
flights_pivot = flights.pivot("month", "year", "passengers")

# Plotting the heatmap
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Heatmap of Flights Data')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```

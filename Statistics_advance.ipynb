Q1: What is the Probability Density Function (PDF)?

The Probability Density Function (PDF) is a function that describes the likelihood of a continuous random variable taking on a particular value. The PDF must satisfy two conditions:

1. The function is non-negative for all values.
2. The integral of the PDF over the entire space is equal to 1.


Q2: What are the types of Probability Distribution?


There are two main types of probability distributions:

1. Discrete Probability Distributions: These distributions apply to discrete random variables (e.g., Binomial, Poisson).
2. Continuous Probability Distributions: These distributions apply to continuous random variables (e.g., Normal, Exponential).


Q3: Python Code for PDF of a Normal Distribution

```python
import numpy as np  
import matplotlib.pyplot as plt  
from scipy.stats import norm  

mean = 0  # Example mean  
std_dev = 1  # Example standard deviation  
x = np.linspace(-5, 5, 100)  
pdf = norm.pdf(x, mean, std_dev)  

plt.plot(x, pdf)  
plt.title('Probability Density Function of Normal Distribution')  
plt.xlabel('x')  
plt.ylabel('PDF')  
plt.grid()  
plt.show()
```

Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4
and plot a histogram of the results using matplotlib.

```python
import numpy as np  
import matplotlib.pyplot as plt  

n = 1000  # Sample size  
p = 0.4  # Probability of success  
samples = np.random.binomial(n=10, p=p, size=n)  

plt.hist(samples, bins=30, density=True, alpha=0.6, color='g')  
plt.title('Histogram of Binomial Distribution')  
plt.xlabel('Number of Successes')  
plt.ylabel('Frequency')  
plt.grid()  
plt.show()
```

Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution
with given mean at a given point.

```python
import numpy as np  
import matplotlib.pyplot as plt  
from scipy.stats import poisson  

lambda_param = 5  # Example mean  
x = np.arange(0, 20)  
cdf = poisson.cdf(x, lambda_param)  

plt.plot(x, cdf, marker='o')  
plt.title('Cumulative Distribution Function of Poisson Distribution')  
plt.xlabel('k')  
plt.ylabel('CDF')  
plt.grid()  
plt.show()
```

Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the
sample mean and variance.

```python
lambda_param = 5  # Example mean  
sample_size = 5  
samples = np.random.poisson(lam=lambda_param, size=sample_size)  

mean = np.mean(samples)  
variance = np.var(samples)  

print("Sample:", samples)  
print("Mean:", mean)  
print("Variance:", variance)
```

Q10. In normal distribution with respect to mean position, where does the least frequent data appear?

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate data following a normal distribution
mean = 0
std_dev = 1
data = np.random.normal(mean, std_dev, 1000)

# Plotting the normal distribution
sns.histplot(data, kde=True)

# Highlighting the extremes (2 standard deviations from the mean)
extreme_left = mean - 2 * std_dev
extreme_right = mean + 2 * std_dev

plt.axvline(extreme_left, color='r', linestyle='--', label='Extreme Left')
plt.axvline(extreme_right, color='r', linestyle='--', label='Extreme Right')

plt.title('Normal Distribution with Extremes Highlighted')
plt.xlabel('Data points')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
